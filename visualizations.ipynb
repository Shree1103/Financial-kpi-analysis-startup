```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("startup_financials_dataset.xlsx", sheet_name="Sheet1")

# Convert date column
df['Only Date'] = pd.to_datetime(df['Only Date'])
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Only Date'], df['Revenue (₹)'], label='Revenue (₹)', color='green', marker='o')
plt.plot(df['Only Date'], df['Operating Expenses (₹)'], label='Expenses (₹)', color='red', marker='o')
plt.title("Revenue vs Expenses Over Time")
plt.xlabel("Month")
plt.ylabel("₹")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Only Date'], df['CAC (₹)'], label='CAC (₹)', color='blue', marker='x')
plt.plot(df['Only Date'], df['LTV (₹)'], label='LTV (₹)', color='orange', marker='x')
plt.title("CAC vs LTV Over Time")
plt.xlabel("Month")
plt.ylabel("₹")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar for Burn Rate
ax1.bar(df['Only Date'], df['Burn Rate (₹)'], color='crimson', label='Burn Rate (₹)')
ax1.set_ylabel("Burn Rate (₹)", color='crimson')
ax1.tick_params(axis='y', labelcolor='crimson')

# Line for Runway
ax2 = ax1.twinx()
ax2.plot(df['Only Date'], df['Runway (Months)'], color='navy', marker='o', label='Runway (Months)')
ax2.set_ylabel("Runway (Months)", color='navy')
ax2.tick_params(axis='y', labelcolor='navy')

plt.title("Burn Rate and Runway Over Time")
fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Only Date', y='Total Customers', data=df, marker='o', color='teal')
plt.title("Customer Growth Over Time")
plt.xlabel("Month")
plt.ylabel("Total Customers")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['LTV:CAC Ratio'] = df['LTV (₹)'] / df['CAC (₹)']
plt.figure(figsize=(10, 5))
sns.lineplot(x='Only Date', y='LTV:CAC Ratio', data=df, marker='o', color='purple')
plt.title("LTV:CAC Ratio Over Time")
plt.axhline(y=3, color='gray', linestyle='--', label='Ideal Threshold (3:1)')
plt.xlabel("Month")
plt.ylabel("Ratio")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
